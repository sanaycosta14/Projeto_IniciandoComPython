```python
clientes = [ ]
contas = { }

def cadastro ( ):
  nome = input('Digite seu nome:')
  cpf = input('Digite seu CPF: ')
  email= input ('Digite seu e-mail: ')
  senha = input ('Digite uma senha: ')

  cliente = {'nome': nome, 'cpf': cpf, 'email': email, 'senha': senha}
  clientes.append (cliente)
  contas [cpf] = {'saldo': 0}
  
  print(f"Cliente {nome} cadastrado com sucesso!".upper())

def lista ( ):
  print("-"*27)
  print ('== CLIENTES CADASTRADOs ==')
  if len(clientes) > 0:
    for cliente in clientes:
      saldo = contas [cliente['cpf']]['saldo']
      print(f"Nome: {cliente['nome']},\nCPF: {cliente['cpf']},\nEmail: {cliente['email']},\nSaldo: R$ {saldo:.2f}")
  else:
    print ("Nenhum cliente cadastrado!!!!".upper())

def atualizar ( ):
  cpf = input ('Digite o CPF do cliente a ser atualizado: ')
  cliente = buscar (cpf)
  if cliente is not None:
    nome = input ("Digite seu nome: ")
    email = input ("Digite seu email: ")
    senha = input ("Digite seu senha: ")

    cliente ['nome'] = nome
    cliente ['email'] = email
    cliente ['senha'] = senha

    print (f"Dados do cliente {cpf} atualizado com sucesso!".upper())
  else:
    print ('Cliente não encontrado!'.upper())

def excluir ( ):
  cpf = input ("Cliente digite o seu CPF para prosseguir com a exclusão: ")
  cliente = buscar (cpf)
  if cliente is not None:
    clientes.remove(cliente)
    contas.pop(cpf)
    print ("Cliente excluído com sucesso!.".upper())
  else:
    print ("Cliente não encontrado no banco de Dados!".upper())

def buscar ( cpf): 
  for cliente in clientes:
    if cliente ['cpf'] == cpf:
      return cliente
  return None

while True:
  print ("_"*60)
  print ('=== Banco ClothesBank - Cadastro de clientes e Contas ===')
  print ('1. Cadastrar')
  print ('2. Listar Clientes')
  print ('3. Atualizar cadastro')
  print ('4. Exclusão de cliente')
  print ("5. Encerrar Cessão")
  
  opcao = input ("Digite o número do atendimento que deseja: ")

  if opcao == '1':
    cadastro ( )
  elif opcao == '2':
    lista ( )
  elif opcao == '3':
    atualizar ( )
  elif opcao == '4':
    excluir ( )
  elif opcao == '5':
    print ('Encerrando  a sessão. . .'.upper())
    break
  else:
    print ("Opção inválida, por favor tente novamente".upper())
print ('Sessão encerrada.'.upper())
```
